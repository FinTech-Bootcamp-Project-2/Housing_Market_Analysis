```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.preprocessing.sequence import TimeseriesGenerator
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import tensorflow as tf

mpl.rcParams['figure.figsize'] = (10, 8)
mpl.rcParams['axes.grid'] = False
```

```python
from numpy.random import seed
seed(1)
from tensorflow import random
random.set_seed(2)
```

```python
df = pd.read_csv("cleaned_data.csv", index_col=0, parse_dates=True, infer_datetime_format=True)
df.info()
```

```python
df.plot(subplots=True, figsize=(20,10))
```

```python
df.corr()
```

```python
df_input = df[['Mortgage Rate', 'inflation_common', 'unemployment_rate', '5_year_bond_benchmark', 'Bank of Canada Target Overnight Rate']]
df_input.head()
```

```python
df_input.plot(subplots=True)
```

```python
df_input.describe()
```

```python
scaler=MinMaxScaler()

data_scaled = scaler.fit_transform(df_input)
```

```python
features = data_scaled
target = data_scaled[:,0]
```

```python
TimeseriesGenerator(features, target, length=2, sampling_rate=1, batch_size=1)[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=123, shuffle=False)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
win_length = 5
batch_size = 32
num_features =5

train_generator = TimeseriesGenerator(X_train, y_train, length=win_length, sampling_rate=1, batch_size=batch_size)
test_generator = TimeseriesGenerator(X_test, y_test, length=win_length, sampling_rate=1, batch_size=batch_size)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.LSTM(128, input_shape= (win_length, num_features), return_sequences=True))
model.add(tf.keras.layers.LeakyReLU(alpha=0.5))
model.add(tf.keras.layers.LSTM(128, return_sequences=True))
model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.LSTM(64, return_sequences=False))
model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.Dense(1))
```

```python
model.summary()
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, mode="min")

model.compile(loss=tf.losses.MeanSquaredError(),
              optimizer='adam',
              metrics=[tf.metrics.MeanAbsoluteError()])

history = model.fit(train_generator, epochs=50,
                              validation_data =test_generator,
                              shuffle=False,
                              callbacks=[early_stopping])
              
```

```python
model.evaluate(test_generator, verbose=0)
```
